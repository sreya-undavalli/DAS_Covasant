**Matches Won By Federer**

```python


df['Double_faults']=df['player1 double faults']/df['player1 total points total']
Double_faults=df['Double_faults']
print(Double_faults)
```

**Average Win**

```python
Avg_wps=df[df['winner']=='Roger Federer'].groupby('surface').size()
print(Avg_wps)
```

```python
df.columns
```

```python
# Create a new column for the proportion of double faults
df['double_faults_proportion'] = df['player1 double faults'] / df['player1 total points total']

# Display simple stats
print(df['double_faults_proportion'].describe())
```

```python
player = 'Roger Federer'
total_matches = len(df)
won_matches = df[df['winner'] == player].shape[0]

win_percentage = (won_matches / total_matches) * 100
print(f"{player} won {win_percentage:.2f}% of matches.")
```

```python
avg_wins_per_surface = df[df['winner'] == player].groupby('surface').size() / df.groupby('surface').size()
print(avg_wins_per_surface)
```

```python
df['start date'] = pd.to_datetime(df['start date'], errors='coerce')
df['year'] = df['start date'].dt.year

# Group by year and calculate average double fault proportion
avg_df_by_year = df.groupby('year')['double_faults_proportion'].mean()
print(avg_df_by_year)
```

```python
import matplotlib.pyplot as plt

avg_df_by_year.plot(kind='line', marker='o', title='Avg Double Fault Proportion by Year')
plt.xlabel('Year')
plt.ylabel('Avg Double Fault Proportion')
plt.grid(True)
plt.show()
```
