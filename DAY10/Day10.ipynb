```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
bike = pd.read_csv(r"code/data/bikes.csv")
```

```python
bike['Date']=pd.to_datetime(bike['Date'], format='%d/%m/%Y')
plt.figure(figsize=(10,6))
plt.plot(bike['Date'],bike['Berri1'], label = 'Berri1')
plt.plot(bike['Date'],bike['PierDup'],label = 'PierDup')
plt.xlabel('Date')
plt.ylabel('Attendance')
plt.title('Daily Attendence of Berri1 & PierDup')
plt.show()
```

```python
weekday_sum=bike.groupby('Weekday').sum(numeric_only=True)
print(weekday_sum)
```

```python
weekday_sum.plot(kind = 'bar',figsize=(10,10))
plt.title('Sum of Attendance')
plt.xlabel('Weekday')
plt.ylabel('Total Attendance')
plt.show()
```
